```python
'On Loan'.split('On Loan')
```

```python
def contract_split(row):
  if '~' in row:
    start_date,end_date = map(float,row.split(' ~ '))
    return start_date,end_date,'Contract'
  else:
    if 'On Loan' in row:
      return np.nan,np.nan,'On Loan'
    return np.nan,np.nan,row
```

```python
x = df['contract'].apply(lambda x: contract_split(x))
```

```python
new_columns = ['contract_date_start','contract_date_end','contract_type']
df[new_columns] = list(x)
```

### Height and Weight

```python
df['height'].unique()
```

```python
def to_cm(row):
  if 'cm' in row:
    return float(row.replace('cm',''))
  else:
    feet,inches = row.split('\'')
    inches = inches.replace('"','')
    return int(feet)*30.48+int(inches)*2.54
```

```python
df['height'] = df['height'].apply(to_cm)
```

```python
df['weight'].unique()
```

```python
def to_kg(row):
  if 'kg' in row:
    return float(row.replace('kg',''))
  else:
    return float(row.replace('lbs',''))*0.45359237
```

```python
df['weight'] = df['weight'].apply(to_kg)
```

### Value,Wage and Release

```python
df['value'].unique()
```

```python
def to_num(row):
  row = str(row)
  if '€' in row:
    row = row.replace('€','')
  if 'K' in row:
    return float(row.replace('K',''))*1000
  elif 'M' in row:
    return float(row.replace('M',''))*1000000
  else:
    try:
       return float(row)
    except:
       return np.nan
```

```python
df['value'] = df['value'].apply(to_num)
```
